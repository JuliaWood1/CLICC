### Picking the columns I want to include in my dataset, because there are ove 500 columns in the dataframes above.

## New start

```python
status = df['Custom field (UCLA Staff/Faculty Status (Non-Surveyed))'].fillna(df['Custom field (UCLA Student Status (Non-Surveyed))'])
status
df['UCLA Status'] = status
df.head(30)  
sum(df['UCLA Status'].isna())
df = df.drop(columns = ['Custom field (UCLA Staff/Faculty Status (Non-Surveyed))', 'Custom field (UCLA Student Status (Non-Surveyed))'])
```

```python
df.head()
```

```python
df = df.groupby(["UCLA Status", 'Custom field (UID Number)'])[['Custom field (Assigned Laptop/iPad Equipment)', 'Custom field (Getting a MiFi Hotspot?)' ]].aggregate(list).reset_index()
df.head()

```

```python
mifi = [1 if 'Yes' in entry else 0 for entry in df['Custom field (Getting a MiFi Hotspot?)']]
```

```python
df['Mifi'] = mifi
df = df.drop(columns = ['Custom field (Getting a MiFi Hotspot?)'])
```

```python
df['Custom field (Assigned Laptop/iPad Equipment)'].value_counts()
```

```python
ChromeBook = [1 if 'CLICC ChromeBook' in entry else 0 for entry in df['Custom field (Assigned Laptop/iPad Equipment)']]
ChromeBook
df['CLICC ChromeBook'] = ChromeBook
```

```python
HPLaptop = [1 if 'CLICC HP Laptop' in entry else 0 for entry in df['Custom field (Assigned Laptop/iPad Equipment)']]
HPLaptop
df['CLICC HP Laptop'] = HPLaptop
```

```python
CMacBook = [1 if 'CLICC MacBook' in entry else 0 for entry in df['Custom field (Assigned Laptop/iPad Equipment)']]
CMacBook
df['CLICC MacBook'] = CMacBook
```

```python
ITSMacBook = [1 if 'ITS MacBook' in entry else 0 for entry in df['Custom field (Assigned Laptop/iPad Equipment)']]
ITSMacBook
df['ITS MacBook'] = ITSMacBook
```

```python
ITSTouchscreenLaptop = [1 if 'ITS Touchscreen Laptop' in entry else 0 for entry in df['Custom field (Assigned Laptop/iPad Equipment)']]
ITSTouchscreenLaptop
df['ITS Touchscreen Laptop'] = ITSTouchscreenLaptop
```

```python
ITSWindowsLaptop = [1 if 'ITS Windows Laptop' in entry else 0 for entry in df['Custom field (Assigned Laptop/iPad Equipment)']]
ITSWindowsLaptop
df['ITS Windows Laptop'] = ITSWindowsLaptop
```

```python
ITSiPad = [1 if 'ITS iPad' in entry else 0 for entry in df['Custom field (Assigned Laptop/iPad Equipment)']]
ITSiPad
df['ITS iPad'] = ITSiPad
```

```python
df = df.drop(columns = ['Custom field (Assigned Laptop/iPad Equipment)'])
df.head()
```

```python
df['Custom field (UID Number)'] = df['Custom field (UID Number)'].str.replace(' ', '')
```
